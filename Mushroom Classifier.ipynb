```python
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("datasets_478_974_mushrooms.csv")
pd.pandas.set_option("display.max_columns",None)
df
```

```python
fet_na=[fet for fet in df.columns if (df[fet]=="?").sum()>1]
for fet in fet_na:
    print(fet,np.round((df[fet]=="?").mean(),4))
```

```python
#df=df.drop("stalk-root",axis=1)
df.columns
```

```python
#for fet in df.columns:
df["result"]=np.where(df["class"]=="e",1,0)
df=df.drop(["class"],axis=1)
df
```

```python
df["cap-shape"]=df["cap-shape"].map({"b":0,"c":1,"x":2,"f":3,"k":4,"s":5})
df["cap-surface"]=df["cap-surface"].map({"f":0,"g":1,"y":2,"s":3})
df["cap-color"]=df["cap-color"].map({"n":0,"b":1,"c":2,"g":3,"r":4,"p":5,"u":6,"e":7,"w":8,"y":9})
df["bruises"]=df["bruises"].map({"t":1,"f":0})
df["odor"]=df["odor"].map({"a":0,"l":1,"c":2,"y":3,"f":4,"m":5,"n":6,"p":7,"s":8})
df["gill-attachment"]=df["gill-attachment"].map({"a":0, "d":1, "f":2, "n":3})
df["gill-spacing"]=df["gill-spacing"].map({"c":0,"w":1,"d":2})
df["gill-size"]=df["gill-size"].map({"b":0,"n":1})
df["gill-color"]=df["gill-color"].map({"k":0,"n":1,"b":2,"h":3,"g":3, "r":4,"o":5,"p":6,"u":7,"e":8,"w":9,"y":10})

df["habitat"]= df["habitat"].map({'g':0,'l':1,'m':2,'p':3,'u':4,'w':5,'d':6})

df["population"]= df["population"].map({'a':0,'c':1,'n':2,'s':3,'v':4,'y':5})

df["ring-number"]= df["ring-number"].map({'n':0,'o':1,'t':2})

 
df["ring-type"]= df["ring-type"].map({'c':0,'e':1,'f':2,'l':3,'n':4,'p':5,'s':6,'z':7})
 
df["stalk-shape"]= df["stalk-shape"].map({'e':0,'t':1})




 
df["stalk-color-above-ring"]= df["stalk-color-above-ring"].map({'n':0,'b':1,'c':2,'g':3,'o':4,'p':5,'e':6,'w':7,'y':8})
df["stalk-color-below-ring"]= df["stalk-color-below-ring"].map({'n':0,'b':1,'c':2,'g':3,'o':4,'p':5,'e':6,'w':7,'y':8})
df
```

```python
df=df[[ 'cap-shape', 'cap-surface', 'cap-color',  'odor', 'gill-color',
       'gill-attachment', 'gill-spacing','habitat', 'population','stalk-color-below-ring', 'ring-number', 'ring-type',
       'stalk-shape',
       'stalk-color-above-ring',  
    
        'gill-size','bruises',"result"]]
df
```

```python
df
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
y=df["result"]

#from sklearn.model_selection import train_test_split
#X_train, X_test, y_train, y_test = train_test_split(df1, y, test_size=0.3, random_state=0)


#log
df=df.drop("result",axis=1)
#df=df.drop("class",axis=1)
log=LogisticRegression()
log.fit(df,y)
a=np.mean(cross_val_score(log,df,y))
a
```

```python
from sklearn.ensemble import ExtraTreesClassifier 
extra_tree_forest = ExtraTreesClassifier(n_estimators = 10, 
                                        criterion ='entropy') 
  
# Training the model 
extra_tree_forest.fit(df, y) 
  
# Computing the importance of each feature 
feature_importance = extra_tree_forest.feature_importances_ 
top=pd.Series(feature_importance,index=df.columns)
top.sort_values(ascending=False)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=7, random_state=0)
clf.fit(df, y)
```

```python
a=np.mean(cross_val_score(clf,df,y))
a
```

```python
pickle.dump(clf, open('model.pkl','wb'))

# Loading model to compare the results


```
